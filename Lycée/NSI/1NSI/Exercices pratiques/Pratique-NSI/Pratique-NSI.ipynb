```python
"Indice du minimum"
def indice_min(nombres):
    min = nombres[0]
    indice_min = 0
    for i in range(len(nombres)):
        if nombres[i] < min:
            min = nombres[i]
            indice_min = i
    return indice_min



# tests

assert indice_min([5]) == 0
assert indice_min([2, 4, 1, 1]) == 2
assert indice_min([5, 3, 2, 5, 2]) == 2
```

```python
"""Remplacer une valeur"""
def remplacer(valeurs, valeur_cible, nouvelle_valeur):
    valeurs_remplacees = []
    for k in valeurs:
        if k == valeur_cible:
            valeurs_remplacees.append(nouvelle_valeur)
        else:
            valeurs_remplacees.append(k)
    return valeurs_remplacees


# Tests
# 1er test
valeurs = [3, 8, 7]
assert remplacer(valeurs, 3, 0) == [0, 8, 7]
assert valeurs == [3, 8, 7]
# 2nd test
valeurs = [3, 8, 3, 5]
assert remplacer(valeurs, 3, 0) == [0, 8, 0, 5]
assert valeurs == [3, 8, 3, 5]
```

```python
"""Maximum"""
def maximum(nombres):
    valeur_max = nombres[0]
    for k in nombres:
        if k > valeur_max:
            valeur_max = k
    return valeur_max


# Tests
assert maximum([98, 12, 104, 23, 131, 9]) == 131
assert maximum([-27, 24, -3, 15]) == 24
```

```python
"Indice première occurence"
def indice(element, tableau):
    indice = None
    for i in range(len(tableau)):
        if tableau[i] == element:
            if indice is None:
                indice = i
    return indice


# tests

assert indice(1, [10, 12, 1, 56]) == 2
assert indice(1, [1, 50, 1]) == 0
assert indice(15, [8, 9, 10, 15]) == 3
assert indice(1, [2, 3, 4]) is None
```

```python
"Dénivelé positif"
def denivele_positif(altitudes):
    denivele_cumule_pos = 0
    for i in range(len(altitudes)-1):
        if altitudes[i+1] - altitudes[i] > 0:
            denivele_cumule_pos += altitudes[i+1] - altitudes[i]
    return denivele_cumule_pos
        




# tests

assert denivele_positif([330, 490, 380, 610, 780, 550]) == 560
assert denivele_positif([200, 300, 100]) == 100
```

```python
"Dernière occurrence"
def derniere_occurrence(tableau, cible):
    derniere_occurrence = None
    for i in range(len(tableau)):
        if tableau[i] == cible:
            derniere_occurrence = i
    if derniere_occurrence is None:
        derniere_occurrence = len(tableau)
    return derniere_occurrence

# tests

assert derniere_occurrence([5, 3], 1) == 2
assert derniere_occurrence([2, 4], 2) == 0
assert derniere_occurrence([2, 3, 5, 2, 4], 2) == 3
```

```python
"Moyenne simple"
def moyenne(valeurs):
    somme = 0
    for k in valeurs:
        somme += k
    return somme/len(valeurs)






# tests

def sont_proches(x, y):
    return abs(x - y) < 10**-6

assert sont_proches(moyenne([10, 20, 30, 40, 60, 110]), 45.0)
assert sont_proches(moyenne([1, 3]), 2.0)
assert sont_proches(moyenne([44, 51, 12, 72, 65, 34]), 46.333333333333336)
```

```python
"Recherche d'indices"
def indices(element, entiers):
    indices = []
    for i in range(len(entiers)):
        if entiers[i] == element:
            indices.append(i)
    return indices


# tests

assert indices(3, [3, 2, 1, 3, 2, 1]) == [0, 3]
assert indices(4, [1, 2, 3]) == []
assert indices(1, [1, 1, 1, 1]) == [0, 1, 2, 3]
assert indices(5, [0, 0, 5]) == [2]
```

```python
"Soleil couchant"
def nb_batiments_eclaires(hauteurs):
    nb_batiments_eclaires = 0
    rayon_soleil_hauteur = 0
    for k in hauteurs:
        if rayon_soleil_hauteur < k:
            nb_batiments_eclaires += 1
            rayon_soleil_hauteur = k
    return nb_batiments_eclaires




# tests

assert 4 == nb_batiments_eclaires([2, 1, 2, 4, 0, 4, 5, 3, 5, 6])
assert 1 == nb_batiments_eclaires([0, 3, 1, 2])
```

```python
"Premier minimum local"
def indice_arret(hauteurs):
    indice_arret = None
    for i in range(len(hauteurs)-1):
        if hauteurs[i] < hauteurs[i+1]:
            if indice_arret is None:
                indice_arret = i
    return indice_arret


# Tests
hauteurs = [3, 2, 5]
assert indice_arret(hauteurs) == 1

hauteurs = [3, 5]
assert indice_arret(hauteurs) == 0

hauteurs = [10, 8, 7, 5, 5, 4, 3, 6, 6, 5, 4, 12]
assert indice_arret(hauteurs) == 6
```

```python
"""Liste des différences"""
def differences(source_1, source_2):
    differences = []
    for k in range(len(source_1)):
        differences.append(not(source_1[k]==source_2[k]))
    return differences



# tests

assert differences([14, 87, 22, 5, 65],
                   [14, 86, 27, 5, 65]) == [False, True, True, False, False]

assert differences([-54], [-54]) == [False]
assert differences([7, 8], [7, 11]) == [False, True]
assert differences([], []) == []
```
